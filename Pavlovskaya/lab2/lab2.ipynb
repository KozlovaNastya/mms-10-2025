```python
from tabulate import tabulate

def Alex(v):
    return 1

def Bob(v):
    return 0

def Clara(preValue):
    return preValue

def Denis(preValue):
    return (preValue + 1) % 2

def Emma(game):
    if game % 20 == 0:
        return 1
    return 0

forFrida = 0
def Frida(preValue):
    global forFrida
    if preValue == 1:
        forFrida = 1
    if forFrida == 1:
        return 1
    return 0

forGeorge = 0
def George(preValue):
    global forGeorge
    if preValue == 1:
        if forGeorge == 1:
            forGeorge = 0
            return 1
        if forGeorge == 0:
            forGeorge = 1
            return 0
    forGeorge = 0
    return 0

def reward(f, s):
    if f == 0 and s == 0:
        return 3, 3
    if f == 0 and s == 1:
        return 0, 5
    if f == 1 and s == 0:
        return 5, 0
    if f == 1 and s == 1:
        return 1, 1

points = {
    Alex: 0,
    Bob: 0, 
    Clara: 0,
    Denis: 0,
    Emma: 0,
    Frida: 0,
    George: 0
}

serie = {
    Alex: 0,
    Bob: 0, 
    Clara: 0,
    Denis: 0,
    Emma: 0,
    Frida: 0,
    George: 0
}

def makeGame(chel1, chel2):
    global points
    global series
    global forFrida
    global forGeorge
    cSer1 = 0
    cSer2 = 0
    points1 = []
    points2 = []
    res1 = 0
    res2 = 0
    fstPar = 0
    secPar = 0
    if chel1 == Denis:
        fstPar = 1
    if chel2 == Denis:
        secPar = 1
    for i in range(200):
        if chel1 == Emma:
            fstPar = i + 1
        fst = chel1(fstPar)
        points1.append(fst)
        if chel2 == Emma:
            secPar = i + 1
        sec = chel2(secPar)
        points2.append(sec)
        rew1, rew2 = reward(fst, sec)
        res1 += rew1
        res2 += rew2

        if rew1 == 5:
            cSer1 += 1
            if cSer1 > serie[chel1]:
                serie[chel1] = cSer1
        else:
            cSer1 = 0
  
        if rew2 == 5:
            cSer2 += 1
            if cSer2 > serie[chel2]:
                serie[chel2] = cSer2
        else:
            cSer2 = 0
        
        fstPar = sec
        secPar = fst
        
    forGeorge = 0
    forFrida = 0
    points[chel1] += res1
    points[chel2] += res2

makeGame(Alex, Alex)
makeGame(Alex, Bob)
makeGame(Alex, Clara)
makeGame(Alex, Denis)
makeGame(Alex, Emma)
makeGame(Alex, Frida)
makeGame(Alex, George)
makeGame(Alex, Alex)
makeGame(Bob, Alex)
makeGame(Clara, Alex)
makeGame(Denis, Alex)
makeGame(Emma, Alex)
makeGame(Frida, Alex)
makeGame(George, Alex)

makeGame(Bob, Bob)
makeGame(Bob, Clara)
makeGame(Bob, Denis)
makeGame(Bob, Emma)
makeGame(Bob, Frida)
makeGame(Bob, George)
makeGame(Bob, Bob)
makeGame(Clara, Bob)
makeGame(Denis, Bob)
makeGame(Emma, Bob)
makeGame(Frida, Bob)
makeGame(George, Bob)

makeGame(Clara, Clara)
makeGame(Clara, Denis)
makeGame(Clara, Emma)
makeGame(Clara, Frida)
makeGame(Clara, George)
makeGame(Clara, Clara)
makeGame(Denis, Clara)
makeGame(Emma, Clara)
makeGame(Frida, Clara)
makeGame(George, Clara)

makeGame(Denis, Denis)
makeGame(Denis, Emma)
makeGame(Denis, Frida)
makeGame(Denis, George)
makeGame(Denis, Denis)
makeGame(Emma, Denis)
makeGame(Frida, Denis)
makeGame(George, Denis)

makeGame(Emma, Emma)
makeGame(Emma, Frida)
makeGame(Emma, George)
makeGame(Emma, Emma)
makeGame(Frida, Emma)
makeGame(George, Emma)

makeGame(Frida, Frida)
makeGame(Frida, George)
makeGame(Frida, Frida)
makeGame(George, Frida)

makeGame(George, George)
makeGame(George, George)

results = []
for player, totalPoints in points.items():
    averagePoints = totalPoints / 14
    results.append([player.__name__, totalPoints, averagePoints])

results.sort(key=lambda x: x[2], reverse=True)

for i, row in enumerate(results, 1):
    row.insert(0, i)

headers = ["месца", "стратэгія", "агульныя ачкі", "сярэднія ачкі"]

print("вынікі па найбольшай набранай колькасці ачкоў")
print(tabulate(results, headers=headers, tablefmt="grid"))
print('\n')

resultsSeries = []
for player, totalSeries in serie.items():
    resultsSeries.append([player.__name__, totalSeries])

resultsSeries.sort(key=lambda x: x[1], reverse=True)

for i, row in enumerate(resultsSeries, 1):
    row.insert(0, i)

headersS = ["месца", "стратэгія", "найбольшая серыя"]

print("вынікі па найдаўжэйшай серыі выйгрышаў (5 ачкоў)")
print(tabulate(resultsSeries, headers=headersS, tablefmt="grid"))
```
