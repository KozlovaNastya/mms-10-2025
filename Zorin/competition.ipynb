```python
rule = [[3,0]
       ,[5,1]]
```

```python
MAX_PLAYERS = 7
class Player:
    free_id = 0
    
    def __init__(self, name = "_NULL"):
        self.name = name
        self.score = [0] * MAX_PLAYERS
        self.streak = [0] * MAX_PLAYERS
        self.globalStreak = 0
        self.id = Player.free_id
        Player.free_id += 1
                
    def act(self, plr):
        return 0;

    def update(self, plr, act):
        pass

    def getTotalScore(self):
        sum = 0
        for score in self.score:
            sum += score
        return sum
```

```python
class Player1(Player):

    def __init__(self, name = "A Player"):
        super().__init__(name)

    def act(self, plr):
        return 1;


class Player0(Player):

    def __init__(self, name = "B Player"):
        super().__init__(name)

    def act(self, plr):
        return 0;


class PlayerCopycat(Player):
    def __init__(self, name = "C Player"):
        super().__init__(name)
        self.memory = [0] * MAX_PLAYERS

    def act(self, plr):
        return self.memory[plr.id];

    def update(self, plr, act):
        self.memory[plr.id] = act


class PlayerNegative(Player):
    def __init__(self, name = "D Player"):
        super().__init__(name)
        self.memory = [1] * MAX_PLAYERS

    def act(self, plr):
        return 1 - self.memory[plr.id];

    def update(self, plr, act):
        self.memory[plr.id] = act


class PlayerMercy20(Player):
    def __init__(self, name = "E Player"):
        super().__init__(name)
        self.step = [0] * MAX_PLAYERS

    def act(self, plr):
        if self.step[plr.id] % 20 == 0:
            return 1
        else:
            return 0

    def update(self, plr, act):
        self.step[plr.id] += 1


class PlayerFlag(Player):
    def __init__(self, name = "F Player"):
        super().__init__(name)
        self.flag = [0] * MAX_PLAYERS

    def act(self, plr):
        return self.flag[plr.id]

    def update(self, plr, act):
        if act == 1:
            self.flag[plr.id] = 1


class PlayerAlt(Player):
    def __init__(self, name = "G Player"):
        super().__init__(name)

    def act(self, plr):
        if plr.score[self.id] > self.score[plr.id]:
            return 0
        else:
            return 1
```

```python
def compete(plrA, plrB):
    actA = plrA.act(plrB)
    actB = plrB.act(plrA)

    plrA.score[plrB.id] += rule[actA][actB]
    if rule[actA][actB] == 5:
        plrA.streak[plrB.id]+=1
    else:
        plrA.streak[plrB.id] = 0
    plrA.globalStreak = max(plrA.streak[plrB.id], plrA.globalStreak)
    
    plrB.score[plrA.id] += rule[actB][actA]
    if rule[actB][actA] == 5:
        plrB.streak[plrA.id]+=1
    else:
        plrB.streak[plrA.id] = 0
    plrB.globalStreak = max(plrB.streak[plrA.id], plrB.globalStreak)
    
    plrA.update(plrB, actB)
    plrB.update(plrA, actA)
    
```

```python
Player.free_id = 0
players = [
            Player1("Alex"),
            Player0("Bob"),
            PlayerCopycat("Clara"),
            PlayerNegative("Denis"),
            PlayerMercy20("Emma"),
            PlayerFlag("Frida"),
            PlayerAlt("George")
    
          ]
plrCnt = len(players)
```

```python
for round in range(200):
    for i in range(plrCnt):
        for j in range(i+1, plrCnt):
            compete(players[i], players[j])
```

```python
for plr in players:
    print (f"{plr.name}, sum - {plr.getTotalScore()}, score - {plr.score}, streak - {plr.globalStreak}")
```
